#### **This data is collated from https://data.gov.in. It has state-wise and district level data on the various crimes committed against women between 2001 to 2015.**

#### **Crimes that are include are :**
* **States and Union Territories**
* **District**
* **Year**
* **'Rape'**
* **'Kidnapping and Abduction'**
* **'Dowry Deaths'**
* **'Assault on women with intent to outrage her modesty'**
* **'Insult to modesty of Women','Cruelty by Husband or his Relatives'**
* **'Importation of Girls'**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
crimes = pd.read_csv('Crime Against Women 2001 to 2015')
```

## 1. Year-Wise Analysis: Trends and Patterns in Crimes Against Women (2001-2015)

```python
yearly_crimes = crimes.groupby('Year')[['Rape', 'Kidnapping & Abduction',
       'Dowry Deaths', 'Assault on women with intent to outrage her modesty',
       'Insult to modesty of Women', 'Cruelty by Husband or his Relatives',
       'Importation of Girls', 'Total Crimes']].sum()
```

```python
yearly_crimes
```

```python
plt.figure(figsize=(10,6),dpi=100)
ax=plt.axes()
ax.spines['top'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['right'].set_visible(False)

ax.grid(linestyle="--",axis="y",color='gray')

sns.barplot(data=yearly_crimes,x=yearly_crimes.index,y='Total Crimes')

for index,value in enumerate(yearly_crimes['Total Crimes']):
    plt.text(index,value,f'{value}',fontsize = 10,ha = 'center',va = 'bottom')
ax.set_xticklabels(ax.get_xticklabels(),rotation =90)

plt.title("Total number of crimes Each year",fontweight='bold',fontsize=15)

plt.xticks(fontweight='bold')
plt.yticks(fontweight='bold')
plt.xlabel("Year",fontweight='bold')
plt.ylabel("Total Crimes",fontweight='bold')
```

```python
# creat dataset which do not include 'Total Crimes' Column
crimes_data = yearly_crimes[['Rape', 'Kidnapping & Abduction',
       'Dowry Deaths', 'Assault on women with intent to outrage her modesty',
       'Insult to modesty of Women', 'Cruelty by Husband or his Relatives',
       'Importation of Girls']]
crimes_data
```

```python
plt.figure(figsize=(10,6),dpi=100)
ax=plt.axes()
ax.spines['top'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['right'].set_visible(False)

ax.grid(linestyle="--",axis="y",color='gray')

sns.lineplot(data=crimes_data)

plt.title("Crime rate against women in India",fontweight='bold',fontsize=15)

plt.xticks(fontweight='bold')
plt.yticks(fontweight='bold')
plt.xlabel("Year",fontweight='bold')
plt.ylabel("Total Crimes",fontweight='bold')
```

```python
plt.figure(figsize = (10,6))
crimes_data.plot.bar(stacked = True,figsize = (10,6))

plt.title("Breakdown of Crime Incidents by Category in India",fontweight='bold',fontsize=15)
plt.grid(True,axis = 'y',color = 'grey',linestyle = '--')
```

```python
df_crimes_data = crimes_data.reset_index()
df_melted = pd.melt(df_crimes_data , id_vars= 'Year' , value_vars = df_crimes_data.columns , var_name= 'Crime_Type', value_name='Total')
# Set the figure size and resolution
plt.figure(figsize=(14, 8), dpi=100)

# Create the grouped bar plot
sns.barplot(x='Year', y='Total', hue='Crime_Type', data=df_melted)

plt.grid(True,axis = 'y',color = 'grey',linestyle = '--')
# Customize the plot
plt.xticks(rotation=45, fontweight='bold')
plt.yticks(fontweight='bold')
plt.xlabel("Year", fontweight='bold')
plt.ylabel("Total Crimes", fontweight='bold')
plt.title("Comparative Analysis of Crime types agaist women",fontweight='bold',fontsize=15)
plt.legend(title="Crime Type")
plt.tight_layout()  # Adjust layout to fit labels and legend

# Display the plot
plt.show()
```

```python
def bar_plot(dataframe,y):
    plt.figure(figsize=(10,6),dpi=100)
    ax=plt.axes()

    ax.grid(linestyle="--",axis="y",color='gray')

    sns.barplot(data = dataframe, x = dataframe.index, y = y)

    # Adding number of crimes on top of each bar
    for index, value in enumerate(dataframe[y]):
        ax.text(index, value, f'{value}', ha='center', va='bottom', fontweight='bold')

    ax.set_xticklabels(ax.get_xticklabels(),rotation =90)
    plt.xticks(fontweight='bold')
    plt.yticks(fontweight='bold')
    plt.xlabel("Sate/Ut",fontweight='bold')
    plt.ylabel(f"Total {y} ",fontweight='bold')
```

```python
def pie_chart(dataframe,x):
    plt.figure(figsize = (10,6))
    plt.pie(dataframe[x], labels = dataframe.index , autopct = '%1.1f%%' )
    plt.show()  
```

```python
def top_5_list(y):
    df=state_crimes.sort_values(by = y , ascending = False)
    first_five = df.head()
    dfr = df.iloc[5:,:]
    # CREATING DATAFRAME SINGLE ROW DATAFRAME OF ALL THE ROWS EXCEPT HEAD()
    dfr = pd.DataFrame(dfr.sum(axis = 0)).transpose()
    
    dfr['Crimes'] ="Others" # reate a columns as ' Crimes'
    
    dfr.set_index('Crimes',inplace = True)
    
    # concat head and others dataframe
    five_n_others = pd.concat([first_five,dfr])
    return five_n_others
top_5_list('Importation of Girls')
```

```python
def top_5(y):
    df=state_crimes.sort_values(by = y , ascending = False)
    first_five = df.head()
    dfr = df.iloc[5:,:]
    # CREATING DATAFRAME SINGLE ROW DATAFRAME OF ALL THE ROWS EXCEPT HEAD()
    dfr = pd.DataFrame(dfr.sum(axis = 0)).transpose()
    
    dfr['Crimes'] ="Others" # reate a columns as ' Crimes'
    
    dfr.set_index('Crimes',inplace = True)
    
    # concat head and others dataframe
    five_n_others = pd.concat([first_five,dfr])
    
    


    # Create a figure with 1 row and 2 columns of subplots
    f, ax = plt.subplots(nrows=1, ncols=2, figsize=(12, 5))
    f.patch.set_facecolor('#F2EDD7FF')
    
    # Bar plot
    
    ax[0].bar(first_five.index, first_five[y])
    ax[0].set_title(f'Bar Plot of {y}')
    ax[0].set_xlabel('Categories')
    ax[0].set_ylabel('Values')
    # Adding number of crimes on top of each bar
    for index, value in enumerate(first_five[y]):
         ax[0].text(index, value, f'{value}', ha='center', va='bottom', fontweight='bold')
    
    # Pie chart
    ax[1].pie(five_n_others[y], labels=five_n_others.index, autopct='%1.1f%%')
    ax[1].set_title(f'Pie Chart of {y}')
    
    # Show plot
    plt.tight_layout()
    plt.show()
```
